## Read Data 

```python
import os
from os import listdir
from os.path import isfile, join
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_data(path):
    onlyfiles = [f for f in listdir(path) if isfile(join(path, f) and "jpg" in f or "JPG" in f)]
    data = [np.array(Image.open(path + '/'+ file).resize((150,150))) for file in onlyfiles]
    
    target = []
    for file in onlyfiles:
        if "No" in file:
            target.append(0)
        else:
            target.append(1)
    return np.array(data) / 255. , target
```

```python
train_X, train_y = read_data("./Train")
```

```python
plt.hist(train_y)
```

We divide our data in two parts (train and validation)

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=2,test_size=0.2, random_state=42)
for train_index, test_index in split.split(train_X, train_y):
    X_train = train_X[train_index]
    y_train = np.array(train_y)[train_index]
    X_val = train_X[test_index]
    y_val = np.array(train_y)[test_index]
```

## Model creation

```python
# PyTorch libraries and modules
import torch
from torch.autograd import Variable
from torch.nn import Linear, ReLU, CrossEntropyLoss, Sequential, Conv2d, MaxPool2d, Module, Softmax, BatchNorm2d, Dropout
from torch.optim import Adam, SGD
```

```python
''' This class contains one layer of 4 conv neurons, a layer of BatchNormalization, We use ReLu activation and a max pooling process 2x2 to reduce before the linear layer.'''


class Net(Module):   
    def __init__(self):
        super(Net, self).__init__()

        self.cnn_layers = Sequential(
            # Defining a 2D convolution layer
            Conv2d(3, 4, kernel_size=3, stride=1, padding=1),
            BatchNorm2d(4),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=2, stride=2),
            # Defining another 2D convolution layer
            Conv2d(4, 4, kernel_size=3, stride=1, padding=1),
            BatchNorm2d(4),
            ReLU(inplace=True),
            MaxPool2d(kernel_size=2, stride=2),
        )

        self.linear_layers = Sequential(
            Linear(4 * 37 * 37, 2)
        )

    # Defining the forward pass    
    def forward(self, x):
        x = self.cnn_layers(x)
        x = x.view(x.size(0), -1)
        x = self.linear_layers(x)
        return x
```

```python
# defining the model
model = Net()
# defining the optimizer
optimizer = Adam(model.parameters(), lr=0.01)
# defining the loss function
criterion = CrossEntropyLoss()
print(model)
```

```python
def preprocess_data(X,y):
    X  = torch.from_numpy(X)

    # converting the target into torch format
    y = np.array(y).astype(int);
    y = torch.from_numpy(y)
    X = X.permute(0,3,1,2)
    
    return X,y
```

```python
def train(epoch):
    model.train()
    tr_loss = 0
    # getting the training set
    x_train, train_y = Variable(X_train), Variable(y_train)
    # getting the validation set
    x_val, val_y = Variable(X_val), Variable(y_val)
    
    x_train = x_train.float()
    x_val = x_val.float()
    train_y = train_y.long()
    val_y = val_y.long()
   

    # clearing the Gradients of the model parameters
    optimizer.zero_grad()
    
    # prediction for training and validation set
    output_train = model(x_train)
    output_val = model(x_val)

    # computing the training and validation loss
    loss_train = criterion(output_train, train_y)
    loss_val = criterion(output_val, val_y)
    train_losses.append(loss_train)
    val_losses.append(loss_val)

    # computing the updated weights of all the model parameters
    loss_train.backward()
    optimizer.step()
    tr_loss = loss_train.item()
    if epoch%2 == 0:
        # printing the validation loss
        print('Epoch : ',epoch+1, '\t', 'loss :', loss_val)
```

```python
X_train.shape, X_val.shape
```

## Training process

```python
# defining the number of epochs
n_epochs = 50
# empty list to store training losses
train_losses = []
# empty list to store validation losses
val_losses = []
# training the model

X_train, y_train = preprocess_data(X_train, y_train)
X_val, y_val = preprocess_data(X_val, y_val)
for epoch in range(n_epochs):
    train(epoch)
```

## Model Evaluation

```python
def predict(X,y):

    # prediction for training set
    with torch.no_grad():
        output = model(X.float())
    softmax = torch.exp(output)
    prob = list(softmax.numpy())
    predictions = np.argmax(prob, axis=1)
    
    return predictions
```

```python
from sklearn.metrics import accuracy_score
```
